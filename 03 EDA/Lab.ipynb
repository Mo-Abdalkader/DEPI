#### Sex

```python
# sex

print("Unique Values in Sex Attribute :", data['sex'].unique())
```

**Q:** Write a Python code to convert **Sex** attribute where male to 0 and female to 1 using sex_wrapper dictionary .

```python
sex_wrapper = {'male':0, 'female':1}
data.sex = data.sex.replace(sex_wrapper)
data.head()
```

#### Smoker

```python
# smoker
print("Unique Values in Smoker Attribute :", attributes['smoker'].unique())
```

**Q:** Write a Python code to convert **Smoker** attribute where No to 0 and Yes to 1 using smoker_wrapper dictionary .

```python
smoker_wrapper = {'no':0, 'yes':1}
data.smoker = data.smoker.replace(smoker_wrapper)
data.head()
```

#### Region

```python
# region 
print("Unique Values in Region Attribute :", attributes['region'].unique())
```

As we saw we have distinct 4 values for Region attribute so How we will handle it ?
One-hot encodding what is that? <br>
simply it creating new 4 columns/features/attributes named after unique values and all these features carries either 0 (if it's NOT the value) 1 (if it carries that value)

**Q:** Write a Python code to One Hot Encode **Region** attribute .

```python
data = pd.get_dummies(data, columns=['region'])
data
```
